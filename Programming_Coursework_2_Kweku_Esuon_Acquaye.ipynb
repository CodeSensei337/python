### **Section 4.1: Implementation**<br> 
A full implementation of k-means is carried out including mathematical operations instead of using a library function such as `scikit` or `statsmodels`. 

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy

%matplotlib inline

print("Libraries loaded.")
```

```python
# Creating pandas dataframe for further analysis
iris_df1 = pd.read_csv('Iris.csv', error_bad_lines=False)
print("Original data loaded.")
```

To preserve integrity of original data file from accidental alteration, analysis would be carried out on a copy.

```python
# Creating copy to preserve original data file
iris_df2 = iris_df1.copy()
print("Copied version created as follows:\n")
iris_df2
```

In the next few steps, preliminary Exploratory Data Analysis (EDA) is carried out:

In the next three steps, the index column (Id) is dropped and the target variable string column converted to numeric classes in order to enable better array calculation and manipulation. The dataframe is then converted to a numpy array.

```python
# Dropping index (id) column
iris_df2.drop('Id',axis=1,inplace=True)
print("Index column dropped.")
```

```python
# Changing target variable column from string to numeric dtype
iris_df2["Species"] = pd.Categorical(iris_df2["Species"])
iris_df2["Species"] = iris_df2["Species"].cat.codes
print("Target column converted to numeric classes 0 - 2.")
```

```python
# Converting df to numpy array
data = iris_df2.values[:, 0:4]
category = iris_df2.values[:, 4]
print("Dataframe converted to numpy array.")
```

The next few steps implements k-means algorithm from 1st principle rather than use of a library:

```python
%%time
# Setting number of clusters
k = 3
# Setting number of training data
n = data.shape[0]
# Setting number of features in the data
c = data.shape[1]

# Generating random centers
mean = np.mean(data, axis = 0)
std = np.std(data, axis = 0)
centroids = np.random.randn(k,c)*std + mean

# Plotting data and centroids
colors=['green', 'orange', 'blue']
for i in range(n):
    plt.scatter(data[i, 0], data[i,1], s=8, color = colors[int(category[i])])
plt.scatter(centroids[:,0], centroids[:,1], marker='*', c='r', s=120)
```

```python
%%time
# Storing previous centroids
centroids_old = np.zeros(centroids.shape)

# Storing new centroids
centroids_new = deepcopy(centroids)

data.shape
clusters = np.zeros(n)
distances = np.zeros((n,k))

error = np.linalg.norm(centroids_new - centroids_old)

# Exiting loop at convergence
while error != 0:
    # Measuring distance to every centroid
    for i in range(k):
        distances[:,i] = np.linalg.norm(data - centroids[i], axis=1)
    # Assigning training data to closest centroid
    clusters = np.argmin(distances, axis = 1)
    
    centroids_old = deepcopy(centroids_new)
    # Calculating mean for every cluster and updating centroid
    for i in range(k):
        centroids_new[i] = np.mean(data[clusters == i], axis=0)
    error = np.linalg.norm(centroids_new - centroids_old)
centroids_new   
```

```python
%%time
# Plotting data and centroids
colors=['green', 'orange', 'blue']
for i in range(n):
    plt.scatter(data[i, 0], data[i,1], s=8, color = colors[int(category[i])])
plt.scatter(centroids_new[:,0], centroids_new[:,1], marker='*', c='r', s=120)
```


### **Section 4.2: Assumptions**<br> 
The k-means method (naive, Lloyd's) is valid under the following assumptions: 
1. Values or members of a cluster are mostly spherically (in 3-dimension) or circularly (in 2-dimension) distributed about a central point.
2. Distribution of values is Gaussian or near Gaussian. For example when values form clusters of concentric circles, k-means is ineffective.
3. Each cluster has approximately the same number of samples or members.   

### **Section 4.3: Limitations**<br> 
k-means algorithms have the following limitations: 
1. The number of clusters k is usually manualy predefined (or predetermined by elbow method) before implementation of the algorithm.
2. Attainment of convergence is dependent on the measure of distance used.
3. The algorithm is sensitive to initial centroid values and may converge on local minima. Variants like k++ can be used to offset this, i.e. seeding.
3. Distance-based similarity measures tend to converge to a constant value as dimensionality increases. Dimensionality reduction methods like spectral clustering and Principal Component Analysis (PCA) can be used to offset this. 
4. Unremoved outliers can drag centroids and distort results. 
5. Clusters of varying density and size can distort results.
 
### **Section 4.4: Complexity of Algorithm**<br> 
The complexity of the above algorithm can be measured by how execution time (or more appropriately computing power with execution time) together with required memory storage space varies with input data size. Algorithm complexity can be described by what is reffered to as the 'big O' notation$^2$, which provides general estimation of algorithm run-time, and also a general measure of scalability.True complexity is however dependent on the number of **operations** that the algorithm ends up performing, i.e. this includes number of iterations in subroutines of the code, 'for' loops, and so on.<br> 
<br> 
Determining the above algorithm's complexity will comprise taking the individual operations of the algorithm and calculating the big O for each operation, I attempt to explain in the following oversimplified steps:<br> 
1. Simple operations which take take the same amount of time to run regardless of the inputs are of time complexity O(1), i.e. constant time. 
The number of rows or columns do not affect the complexity here and therefore is the most scalable.
2. Operations where time increases linearly with the size of the input data, i.e. number of rows, are of complexity O(n). If time increases in linear proportion to the number of columns only and not rows, the complexity is O(d). Time complexity dependent on both number of rows and number of columns is of complexity O(nd). Doubling the number of rows or columns will double computation time, while doubling both will quadruple computation time.
3. Operations where time increases exponentially with input, i.e doubling
number of rows will quadruple the computation, or doubling number of columns will quadruple the computation time, are of complexities O(n$^2$) and O(d$^2$) respectively.
4. All the complexities calculated for the individual operations are added together, constants removed, and the highest order term taken as the overall complexity of the algorithm.
5. Due to the presence of iterations in this algorithm (i.e. 'for' and 'while' loops), its complexity is likely to be of the order of:<br>

<center>$O$($n$$^{(k+1)}$)($d$$^{(k+1)}$)</center><br> 

> where $n$ = number of samples/rows,<br>
> $\space$ $\space$ $\space$ $\space$ $\space$$\space$    $d$ = number of dimensions/columns,<br>
> $\space$ $\space$ $\space$ $\space$ $\space$$\space$    $k$ = number of clusters.
